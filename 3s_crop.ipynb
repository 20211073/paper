```python
for i in files:

    image_list = []
    video_file = root+i # 동영상 파일 경로
    print(video_file)
    cap = cv2.VideoCapture(video_file) # 동영상 캡쳐 객체 생성  ---①
    if cap.isOpened(): # 캡쳐 객체 초기화 확인
        w = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        h = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        while True:
            ret, img = cap.read()      # 다음 프레임 읽기      --- ②
            if ret:                     # 프레임 읽기 정상
                image_list.append(img)
            else:                       # 다음 프레임 읽을 수 없슴,
                break                   # 재생 완료
    else:   
        print("can't open video.")      # 캡쳐 객체 초기화 실패
    print("debug")
    print(len(image_list))
    count = len(image_list)//fps
    print(count)
    for k in range(count):
        if k > 5:
            break
        print(k)
        image_list_save = image_list[fps*k:fps*k+fps]
        #print(len(image_list))
        cap.release()                       # 캡쳐 자원 반납
        out = cv2.VideoWriter(save_root+i.split('.')[0]+"_"+str(k)+".mp4", fourcc, 30, (w, h))
        for k in image_list_save:
            out.write(k)
        out.release()
            #break
```
